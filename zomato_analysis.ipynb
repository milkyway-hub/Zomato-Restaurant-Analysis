```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
users = pd.read_csv("users.csv")
restaurant = pd.read_csv("restaurant.csv")
menu = pd.read_csv("menu.csv")
food = pd.read_csv("food.csv")
orders = pd.read_csv("orders.csv")
```

```python
print("orders:", orders.columns)
print("users:", users.columns)
print("menu:", menu.columns)
print("food:", food.columns)
print("restaurant:", restaurant.columns)
```

```python
orders.drop(columns=["Unnamed: 0"], inplace=True)
users.drop(columns=["Unnamed: 0"], inplace=True)
menu.drop(columns=["Unnamed: 0"], inplace=True)
food.drop(columns=["Unnamed: 0"], inplace=True)
restaurant.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
df = orders.merge(users, on="user_id", how="left")
```

```python
df = df.merge(menu, on="r_id", how="left")
```

```python
df = df.merge(food, on="f_id", how="left")
```

```python
df = df.merge(restaurant, left_on="r_id", right_on="id", how="left")
```

```python
top_food = df['item'].value_counts().head(10)
print(top_food)

top_food.plot(kind='bar', color='orange')
plt.title("Top 10 Selling Food Items")
plt.ylabel("Number of Orders")
plt.xlabel("Food Item")
plt.xticks(rotation=45)
plt.show()
```

```python
veg_count = df['veg_or_non_veg'].value_counts()
print(veg_count)

veg_count.plot(kind='pie', autopct='%1.1f%%', colors=['green','red'])
plt.title("Veg vs Non-Veg Orders")
plt.ylabel("")
plt.show()
```

```python
top_restaurant = df.groupby("name_y")["sales_amount"].sum().sort_values(ascending=False).head(10)
print(top_restaurant)

top_restaurant.plot(kind='bar', color='purple')
plt.title("Top 10 Restaurants by Revenue")
plt.ylabel("Revenue")
plt.xlabel("Restaurant")
plt.xticks(rotation=45)
plt.show()
```

```python
city_revenue = df.groupby("city")["sales_amount"].sum().sort_values(ascending=False)
print(city_revenue)

city_revenue.plot(kind='bar', color='teal')
plt.title("City-wise Revenue")
plt.ylabel("Revenue")
plt.xlabel("City")
plt.xticks(rotation=45)
plt.show()
```

```python
gender_orders = df.groupby("Gender")["sales_amount"].sum()
print(gender_orders)

sns.barplot(x=gender_orders.index, y=gender_orders.values, palette="cool")
plt.title("Revenue by Gender")
plt.ylabel("Revenue")
plt.show()
```

```python
rating_revenue = df.groupby("rating")["sales_amount"].mean().sort_index()
print(rating_revenue)

rating_revenue.plot(kind='line', marker='o', color='blue')
plt.title("Average Revenue vs Restaurant Rating")
plt.xlabel("Rating")
plt.ylabel("Average Revenue")
plt.show()
```

```python
cuisine_revenue = df.groupby("cuisine_x")["sales_amount"].sum().sort_values(ascending=False).head(10)
print(cuisine_revenue)

cuisine_revenue.plot(kind='bar', color='gold')
plt.title("Top 10 Cuisine Types by Revenue")
plt.ylabel("Revenue")
plt.xlabel("Cuisine")
plt.xticks(rotation=45)
plt.show()
```
